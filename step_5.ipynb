```python
import pandas as pd
```

```python
# Read CSV file skipping unnamed columns 0 and 1
df = pd.read_csv('/content/data.csv', usecols=lambda column: column not in ["Unnamed: 0", "Unnamed: 1"])

# Print the DataFrame
print(df)
```

```python
# import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# # Load the dataset (replace 'data1.csv' with your CSV file)
# df = pd.read_csv('data1.csv', usecols=lambda column: column not in ["Unnamed: 0", "Unnamed: 1"])

# Basic summary of the dataset
print(df.head())          # Display the first few rows
print(df.info())          # Information about columns and data types
print(df.describe())      # Summary statistics

# Visualizations
# Pairwise scatter plot
sns.pairplot(df, hue='PreferredCategory')
plt.title("Pairwise Scatter Plot")
plt.show()

# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', bins=20, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Total Purchases vs. Annual Income
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='TotalPurchases', y='AnnualIncome', hue='PreferredCategory')
plt.title("Total Purchases vs. Annual Income")
plt.xlabel("Total Purchases")
plt.ylabel("Annual Income")
plt.show()

# Box plot of PreferredCategory vs. Annual Income
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='PreferredCategory', y='AnnualIncome')
plt.title("Box plot of Preferred Category vs. Annual Income")
plt.xlabel("Preferred Category")
plt.ylabel("Annual Income")
plt.show()
```
